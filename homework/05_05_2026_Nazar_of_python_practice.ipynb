Виведіть всі прості числа від 2 до 50.

**Нагадування:** просте число — це число, яке ділиться тільки на 1 і саме на себе.

> 💡 **Підказка:** Для кожного числа `n` від 2 до 50 перевіряйте, чи має воно дільники від 2 до n-1. Можна використати змінну-прапорець `is_prime = True`.

````python
print("Прості числа від 2 до 50:")

for n in range(2, 51):
    is_prime = True


    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n, end=" ")
````

### Завдання 31. Підрахунок голосних ⭐⭐

Запитайте у користувача слово і підрахуйте кількість голосних літер у ньому.

> 💡 **Підказка:** Створіть лічильник `count = 0`. У циклі перевіряйте: `if letter in "аеиіоуяюєї":` і збільшуйте лічильник.

````python
word = input("Введіть слово: ").lower()
vowels = "аеиіоуяюєї"
count = 0

for letter in word:
    if letter in vowels:
        count += 1

print(f"Кількість голосних літер: {count}")
````

Дано список чисел. Відсортуйте його за зростанням **без використання методу `.sort()` та функції `sorted()`**.

> 💡 **Підказка:** Найпростіший алгоритм — «бульбашкове сортування»: у вкладеному циклі порівнюємо сусідні елементи і міняємо їх місцями, якщо ліворуч стоїть більший. Повторюємо, поки не пройдемо весь список без жодної заміни.

````python
# Ваш код для завдання 56:
````

````python
numbers = [42, 17, 89, 3, 56, 91, 24]
# Ваш код:
n = len(numbers)

for i in range(n):
     swapped = False
     for j in range(0, n - i - 1):

        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
            swapped = True

     if not swapped:
        break

print("Відсортований список:", numbers)

````

### Завдання 60. Аналіз тексту ⭐⭐⭐

Дано рядок з реченням. Виконайте такий аналіз і виведіть результати:
1. Кількість літер (без пробілів)
2. Кількість слів
3. Кількість голосних літер
4. Найдовше слово

> 💡 **Підказка:** Метод `.split()` розбиває рядок на список слів. Наприклад, `'hello world'.split()` повертає `['hello', 'world']`.

````python
text = "програмування на python це цікаво і корисно"

letters_only = text.replace(" ", "")
letters_count = len(letters_only)

words = text.split()
words_count = len(words)

vowels = "аеєиіїоуюяaeiouy"
vowels_count = 0

for char in text.lower():
    if char in vowels:
        vowels_count += 1

longest_word = ""
for word in words:
    clean_word = word.strip(".,!?-\"")
    if len(clean_word) > len(longest_word):
        longest_word = clean_word

print(f"Речення: \"{text}\"\n")
print(f"Кількість літер (без пробілів): {letters_count}")
print(f"Кількість слів: {words_count}")
print(f"Кількість голосних літер: {vowels_count}")
print(f"Найдовше слово: \"{longest_word}\"")
````

### Завдання 61. Магазин: статистика продажів ⭐⭐⭐

Дано список продажів за тиждень (по одному числу на день, починаючи з понеділка).
Виведіть:
1. Загальну виручку за тиждень
2. Середню виручку на день
3. Найкращий день (з найбільшою виручкою) — назвою (Понеділок, Вівторок...)
4. Кількість днів, коли виручка була вищою за середню

> 💡 **Підказка:** Створіть список з назвами днів: `days = ["Понеділок", "Вівторок", ...]`. Індекс найкращого дня знайдіть у циклі (або через `sales.index(max(sales))`).

````python
# Ваш код для завдання 61:
````

````python
sales = [3200, 2800, 4100, 3500, 5200, 7800, 6500]
# Ваш код:
days = ["Понеділок", "Вівторок", "Середа", "Четвер", "П'ятниця", "Субота", "Неділя"]

total = sum(sales)

average = total / len(sales)

best_index = sales.index(max(sales))
best_day = days[best_index]

above_average = 0
for day_sales in sales:
    if day_sales > average:
        above_average += 1

print(f"Загальна виручка за тиждень: {total} грн")
print(f"Середня виручка на день: {average:.2f} грн")
print(f"Найкращий день: {best_day} ({max(sales)} грн)")
print(f"Днів з виручкою вище середньої: {above_average}")
````

Дано два списки: команди та їхні очки (відповідні позиції).
Виведіть турнірну таблицю, ВІДСОРТОВАНУ за спаданням очок, у такому форматі:
```
1. Шахтар — 67 очок
2. Динамо — 64 очки
...
```

> 💡 **Підказка:** Найпростіший спосіб: у циклі знаходьте команду з максимальними очками, виводьте її, потім видаляйте цю пару (команду і очки) з обох списків. Повторюйте, поки списки не спорожніють.

````python
# Ваш код для завдання 62:
````

````python
teams = ["Динамо", "Шахтар", "Зоря", "Дніпро", "Ворскла", "Олександрія"]
points = [64, 67, 52, 48, 39, 45]
# Ваш код:
place = 1

while len(teams) > 0:
    max_index = points.index(max(points))

    p = points[max_index]
    if p % 10 == 1 and p % 100 != 11:
        word = "очко"
    elif 2 <= p % 10 <= 4 and not (12 <= p % 100 <= 14):
        word = "очки"
    else:
        word = "очків"

    print(f"{place}. {teams[max_index]} — {p} {word}")

    teams.pop(max_index)
    points.pop(max_index)

    place += 1
````
