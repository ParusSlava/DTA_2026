### ✏️ Завдання 2.1

Є масив денних продажів магазину за 2 тижні (14 днів):

```python
sales = np.array([320, 415, 290, 480, 350, 620, 580,
                  310, 440, 305, 510, 370, 650, 595])
```

Виконай такі вибірки **без циклів**, лише через зрізи:
1. Продажі першого тижня (дні 1–7)
2. Продажі другого тижня (дні 8–14)
3. Продажі у вихідні: 6-й, 7-й, 13-й, 14-й день — використай індексацію списком `[[5,6,12,13]]`
4. Продажі через день, починаючи з першого (непарні дні: 1, 3, 5...)
5. Масив у зворотному порядку

````python
sales = np.array([320, 415, 290, 480, 350, 620, 580,
                  310, 440, 305, 510, 370, 650, 595])

# Твій код:
# 1. Продажі першого тижня (дні 1–7)
first_week = sales[0:7]

# 2. Продажі другого тижня (дні 8–14)
second_week = sales[7:]

# 3. Продажі у вихідні: 6-й, 7-й, 13-й, 14-й день — використай індексацію списком [[5,6,12,13]]
weekends = sales[[5, 6, 12, 13]]

# 4. Продажі через день, починаючи з першого (непарні дні: 1, 3, 5...)
odd_days = sales[::2]

# 5. Масив у зворотному порядку
reversed_sales = sales[::-1]

print("1. Перший тиждень:               ", first_week)
print("2. Другий тиждень:               ", second_week)
print("3. Вихідні дні:                  ", weekends)
print("4. Через день (непарні дні):     ", odd_days)
print("5. Масив у зворотному порядку:   ", reversed_sales)

````

### ✏️ Завдання 3.1

Є масив цін товарів у доларах. Виконай **без циклів**:

```python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])
```

1. Переведи ціни в гривні (курс: **1 USD = 41.5 UAH**), результат округли до 2 знаків
2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
5. Виведи результати з поясненнями

````python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])

# Твій код:
# 1. Переведи ціни в гривні (курс: 1 USD = 41.5 UAH), результат округли до 2 знаків
prices_uah = np.round(prices_usd * 41.5, 2)

# 2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
prices_vat = np.round(prices_uah * 1.2, 2)

# 3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
min_price = np.min(prices_vat)
max_price = np.max(prices_vat)
mean_price = np.round(np.mean(prices_vat), 2)

# 4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
total_sum = np.round(np.sum(prices_vat), 2)

# 5. Виведи результати з поясненнями
print("1. Ціни в гривнях (без ПДВ):")
print(prices_uah)

print("\n2. Ціни в гривнях (з ПДВ 20%):")
print(prices_vat)

print("\n3. Метрики для цін з ПДВ:")
print(f"   - Мінімальна ціна: {min_price} грн")
print(f"   - Максимальна ціна: {max_price} грн")
print(f"   - Середня ціна товару: {mean_price} грн")

print("\n4. Загальна вартість кошика:")
print(f"   - Разом до сплати: {total_sum} грн")
````

### ✏️ Завдання 3.2

Є дані про кількість товарів на складі та їх собівартість.
Виконай **без циклів**:

```python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']
```

1. Порахуй загальну вартість кожного товару на складі (`quantity * cost_each`)
2. Знайди загальну вартість всього складу
3. Знайди товар з найбільшою сумарною вартістю — використай `np.argmax()`
4. Знайди товар з найменшою сумарною вартістю — `np.argmin()`
5. Виведи середню вартість одного найменування на складі

````python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']

# Твій код:
# 1. Порахуй загальну вартість кожного товару на складі (quantity * cost_each)
total_per_item = quantity * cost_each

# 2. Знайди загальну вартість всього складу
warehouse_total = np.sum(total_per_item)

# 3. Знайди товар з найбільшою сумарною вартістю — використай np.argmax()
max_index = np.argmax(total_per_item)
max_item_name = names[max_index]
max_item_value = total_per_item[max_index]

# 4. Знайди товар з найменшою сумарною вартістю — np.argmin()
min_index = np.argmin(total_per_item)
min_item_name = names[min_index]
min_item_value = total_per_item[min_index]

# 5. Виведи середню вартість одного найменування на складі
average_value = np.mean(total_per_item)

print("1. Сумарна вартість кожного товару на складі:")
for name, total in zip(names, total_per_item):
    print(f"   - {name}: {total:,} грн")

print(f"\n2. Загальна вартість всього складу: {warehouse_total:,} грн")

print(f"\n3. Товар з найбільшою вартістю: {max_item_name} ({max_item_value:,} грн)")

print(f"\n4. Товар з найменшою вартістю: {min_item_name} ({min_item_value:,} грн)")

print(f"\n5. Середня вартість одного найменування: {average_value:,.2f} грн")
````

### ✏️ Завдання 4.1

Є масив часу виконання замовлень у хвилинах (норма — до 30 хв):

```python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])
```

1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)
2. Знайди всі **запізнілі** замовлення (> 30 хв)
3. Порахуй кількість вчасних і запізнілих
4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
5. Яка середня затримка серед запізнілих замовлень?

````python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])

# Твій код:
# 1. Знайди всі замовлення, виконані вчасно (≤ 30 хв)
on_time_mask = delivery_time <= 30
on_time_orders = delivery_time[on_time_mask]

# 2. Знайди всі запізнілі замовлення (> 30 хв)
delayed_mask = delivery_time > 30
delayed_orders = delivery_time[delayed_mask]

# 3. Порахуй кількість вчасних і запізнілих
count_on_time = len(on_time_orders)
count_delayed = len(delayed_orders)
total_orders = len(delivery_time)

# 4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
pct_delayed = np.round((count_delayed / total_orders) * 100, 1)

# 5. Яка середня затримка серед запізнілих замовлень?
delays = delayed_orders - 30
average_delay = np.round(np.mean(delays), 1)

print("1. Замовлення, виконані вчасно (≤ 30 хв):")
print(f"   {on_time_orders}")

print("\n2. Запізнілі замовлення (> 30 хв):")
print(f"   {delayed_orders}")

print("\n3. Статистика кількості:")
print(f"   - Вчасних замовлень: {count_on_time} шт.")
print(f"   - Запізнілих замовлень: {count_delayed} шт.")

print("\n4. Відсоток запізнілих замовлень:")
print(f"   - {pct_delayed}% від усіх замовлень")

print("\n5. Середня затримка серед запізнілих замовлень:")
print(f"   - {average_delay} хв. (на стільки в середньому перетримували кожне проблемне замовлення)")
````

np.random.seed(42)

### ✏️ Завдання 5.1

```python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])
```

1. Перетвори масив у матрицю **6×4** (6 блоків по 4 години) і виведи
2. Знайди суму продажів у кожному 4-годинному блоці (по рядках)
3. Відсортуй вихідний 1D масив за **зростанням** і виведи
4. Знайди **індекси** годин відсортованих за зростанням продажів — `np.argsort()`
5. Виведи топ-3 найпродуктивніших години (їхні номери, рахуючи з 0)

````python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])

# Твій код:
# 1. Перетвори масив у матрицю 6×4 (6 блоків по 4 години) і виведи
matrix_6x4 = hourly.reshape(6, 4)

print(matrix_6x4)
# 2. Знайди суму продажів у кожному 4-годинному блоці (по рядках)
block_sums = np.sum(matrix_6x4, axis=1)

# 3. Відсортуй вихідний 1D масив за зростанням і виведи
sorted_hourly = np.sort(hourly)

# 4. Знайди індекси годин відсортованих за зростанням продажів — np.argsort()
sorted_indices = np.argsort(hourly)

# 5. Виведи топ-3 найпродуктивніших години (їхні номери, рахуючи з 0)
top_3_hours = sorted_indices[:-4:-1]

print("1. Матриця 6х4 (6 блоків по 4 години):")
print(matrix_6x4)

print("\n2. Сума продажів у кожному 4-годинному блоці:")
print(block_sums)

print("\n3. Вихідний масив, відсортований за зростанням:")
print(sorted_hourly)

print("\n4. Індекси годин, відсортовані за зростанням продажів:")
print(sorted_indices)

print("\n5. Топ-3 найпродуктивніших години (номери годин з 0):")
print(top_3_hours)
````

Змоделюй результати маркетингового опитування 50 клієнтів.

1. Встанови `np.random.seed(99)`
2. Згенеруй масив `ages` — вік 50 клієнтів: цілі числа від 18 до 65
3. Згенеруй масив `scores` — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
4. Обріж `scores` так щоб значення були від 1 до 10 — `np.clip(arr, 1, 10)`
5. Виведи:
   - середній вік клієнтів
   - середню оцінку задоволеності
   - кількість клієнтів з оцінкою >= 8 («задоволені»)
   - відсоток задоволених

````python
# Твій код:
# 1. Встанови np.random.seed(99)
np.random.seed(99)

# 2. Згенеруй масив ages — вік 50 клієнтів: цілі числа від 18 до 65
ages = np.random.randint(18, 66, size=50)

# 3. Згенеруй масив scores — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
raw_scores = np.random.normal(loc=7.0, scale=1.5, size=50)
scores_rounded = np.round(raw_scores, 1)

# 4. Обріж scores так щоб значення були від 1 до 10 — np.clip(arr, 1, 10)
scores = np.clip(scores_rounded, 1.0, 10.0)

# 5. Виведи:
     # середній вік клієнтів
     # середню оцінку задоволеності
     # кількість клієнтів з оцінкою >= 8 («задоволені»)
     # відсоток задоволених

average_age = np.mean(ages)
average_score = np.mean(scores)
satisfied_mask = scores >= 8.0
satisfied_count = np.sum(satisfied_mask)
satisfied_percentage = (satisfied_count / len(scores)) * 100


print("--- РЕЗУЛЬТАТИ МАРКЕТИНГОВОГО ОПИТУВАННЯ ---")
print(f"1. Середній вік опитаних клієнтів:      {average_age:.1f} років")
print(f"2. Середня оцінка задоволеності:        {average_score:.2f} з 10")
print(f"3. Кількість задоволених клієнтів (≥8): {satisfied_count} осіб")
print(f"4. Відсоток задоволених клієнтів:       {satisfied_percentage:.1f}%")
````

## 🏆 Фінальний проєкт — Аналіз даних магазину

> **Використовуємо все з тьюторіалу разом**

Ти аналітик даних невеликого магазину електроніки.
Тобі передали масив продажів за 8 тижнів (56 днів).

Виконай повний аналіз та дай відповіді на запитання керівника.

````python
# Дані — не змінюй цю клітинку
np.random.seed(2024)
base = np.random.randint(200, 800, 56)

# Вихідні продаються краще
weekend_boost = np.array([1 if i % 7 in [5, 6] else 0 for i in range(56)])
sales_56 = (base + weekend_boost * np.random.randint(100, 300, 56)).astype(int)

print('Дані отримано. Масив sales_56 готовий до аналізу.')
print(f'Елементів: {sales_56.size}, форма: {sales_56.shape}')
````

### Запитання 1 — Загальна статистика (Розділ 1, 3)
Виведи:
- Загальна виручка за 56 днів
- Середньоденна виручка (округли до 2 знаків)
- Мінімальний та максимальний денний продаж
- Стандартне відхилення (наскільки продажі «скачуть»)

````python
# Твій код:
total_revenue = np.sum(sales_56)
mean_revenue = np.mean(sales_56)
min_sale = np.min(sales_56)
max_sale = np.max(sales_56)
std_dev = np.std(sales_56)
print(f"• Загальна виручка за 56 днів:        {total_revenue:,} грн")
print(f"• Середньоденна виручка:              {mean_revenue:.2f} грн")
print(f"• Мінімальний денний продаж:          {min_sale} грн")
print(f"• Максимальний денний продаж:         {max_sale} грн")
print(f"• Стандартне відхилення (волатильність): {std_dev:.2f} грн")
````

### Запитання 2 — Тижневий аналіз (Розділ 5)
1. Перетвори `sales_56` у матрицю **8×7** (8 тижнів по 7 днів)
2. Знайди виручку кожного тижня (сума по рядках)
3. Знайди найкращий тиждень і виведи його номер (рахуючи з 1) та суму
4. Знайди середньоденну виручку по кожному дню тижня (середнє по стовпцях)
5. Виведи який день тижня в середньому найприбутковіший (0=Пн, 6=Нд)

````python
# Твій код:
matrix_8x7 = sales_56.reshape(8, 7)
weekly_revenue = np.sum(matrix_8x7, axis=1)

best_week_idx = np.argmax(weekly_revenue)
best_week_num = best_week_idx + 1
best_week_sum = weekly_revenue[best_week_idx]

weekday_means = np.mean(matrix_8x7, axis=0)
best_day_idx = np.argmax(weekday_means)

print("• Виручка по тижнях:")
for w, rev in enumerate(weekly_revenue, 1):
    print(f"  - Тиждень {w}: {rev:,} грн")
print(f"• Найкращий тиждень: №{best_week_num} з виручкою {best_week_sum:,} грн")
print("• Середня виручка по днях тижня (0=Пн, 6=Нд):")
days_names = ["Пн", "Вт", "Ср", "Чт", "Пт", "Сб", "Нд"]
for d_idx, d_mean in enumerate(weekday_means):
    print(f"  - {days_names[d_idx]} (Індекс {d_idx}): {d_mean:.2f} грн")
print(f"• Найприбутковіший день тижня: {days_names[best_day_idx]} (індекс {best_day_idx})")
````

### Запитання 3 — Фільтрація та маскування (Розділ 4)
1. Знайди кількість **«слабких» днів** — продажі нижче 400 грн
2. Знайди кількість **«сильних» днів** — продажі вище 700 грн
3. Яка середня виручка у слабкі дні і яка у сильні?
4. Створи масив `labels` за допомогою `np.where()`:
   - `'Слабкий'` якщо < 400
   - `'Сильний'` якщо > 700
   - `'Нормальний'` в решті випадків (підказка: вкладений `np.where`)
5. Виведи перші 14 днів: продажі та їх мітки

````python
# Твій код:
weak_mask = sales_56 < 400
strong_mask = sales_56 > 700

weak_count = np.sum(weak_mask)
strong_count = np.sum(strong_mask)

weak_mean = np.mean(sales_56[weak_mask]) if weak_count > 0 else 0
strong_mean = np.mean(sales_56[strong_mask]) if strong_count > 0 else 0

labels = np.where(sales_56 < 400, 'Слабкий',
                 np.where(sales_56 > 700, 'Сильний', 'Нормальний'))

print(f"• Кількість «слабких» днів (<400 грн): {weak_count} днів")
print(f"• Кількість «сильних» днів (>700 грн): {strong_count} днів")
print(f"• Середня виручка у слабкі дні:        {weak_mean:.2f} грн")
print(f"• Середня виручка у сильні дні:        {strong_mean:.2f} грн")
````
