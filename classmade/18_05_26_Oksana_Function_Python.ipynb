## Типи результатів функцій

````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

````python
# Повертає Non Type, без зміни даних і виводить результат
def info(d):
  print(d)

#info(my_dict)
type(info(my_dict))
````

````python
# Повертає Non Type, змінюж почтакові дані і виводить результат
def info(d):
    d[6] = "six"
    print(d)


#info(my_dict)
type(info(my_dict))

````

````python
# Повертає Non Type, змінюж почтакові дані, НЕ виводить результат
def info(d):
    d[6] = "six"

print(f"my_dict: {my_dict}")
type(info(my_dict))
````

````python
def func_3(a=5, b=7):
  print(a + b)

func_3()  # 5 + 7
func_3(10)  # 10 + 7 (a=10)
func_3(10, 20)  # 10 + 20 (a=10, b=20)
func_3(b=10)  # 5 + 10 (b=10)
````

````python
def func_4(a, b=7):
  print(a + b)

# func_4()  # TypeError: func_4() missing 1 required positional argument: 'a'
func_4(10)  # 10 + 7 (a=10)
func_4(10, 20)  # 10 + 20 (a=10, b=20)
````
