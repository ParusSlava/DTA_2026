## Типи результатів функцій

````python
my_dict = {
    1: "one",
    2: "two",
    3: "three",
    4: "four",
    5: "five"
}
````

````python
# Повертає Non Type, без зміни даних і виводить результат
def info(d):
  print(d)

#info(my_dict)
type(info(my_dict))
````

````python
# Повертає Non Type, змінюж почтакові дані і виводить результат
def info(d):
    d[6] = "six"
    print(d)


#info(my_dict)
type(info(my_dict))

````

````python
# Повертає Non Type, змінюж почтакові дані, НЕ виводить результат
def info(d):
    d[6] = "six"

print(f"my_dict: {my_dict}")
type(info(my_dict))
````
