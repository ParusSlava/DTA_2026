### 📋 Задача 7 — List comprehension
Перепиши кожну операцію **одним рядком** за допомогою list comprehension:
`[вираз for елемент in список]`

````python
prices = [100, 250, 80, 420, 175, 310]

# 1. Список цін з ПДВ (×1.2) — одним рядком
prices_vat = [price * 1.2 for price in prices]
print(f"Список цін з ПДВ: {prices_vat}")

# 2. Тільки ціни більше 200 — одним рядком
big_prices = [el for el in prices if el > 200]
print(f"Ціни більше 200: {big_prices}")

# 3. Рядки виду 'Ціна: 100 грн' для кожної ціни — одним рядком
labels = [f"Ціна: {price} грн" for price in prices]
print(labels)
````
